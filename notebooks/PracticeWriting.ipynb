```python
import pandas as pd
import time
import yfinance as yf
from pathlib import Path
import datetime

indexes = {
    "S&P500" : "^GSPC",
    "NASDAQ" : "^IXIC",
    "DOW JONES" : "^DJI",
    "RUSSELL 2000" : "^RUT"
}

def get_stock_data(indexes:dict, start_year:int=1990, end_year:int=2025, output_dir: str="data"):
    Path("data").mkdir(exist_ok=True)
    for name, symbol in indexes.items():
        for year in range(start_year, end_year + 1):
            file_path = f"{output_dir}/{name.lower().replace(' ','_')}.csv"
            if Path(file_path).exists():
                print(f"Skipping {symbol} for download.")
                continue
            start = f"{year}-01-01"
            end = f"{year}-12-31"
            try:
                print(f"Downloading {symbol} for {start}-{end}")
                df = yf.download(symbol, start=start, end=end)
                if df.empty:
                    print(f"Skipping {name} for {year}")
                    continue
                file_path = f"data/{name.lower().replace(' ','_')}_{year}.csv"
                df.to_csv(file_path)
                print(f"Saved to {file_path}")
                time.sleep(10)
            except Exception as e:
                print(f"Failed to download {name}/{symbol} for {year}: {e}")
```
