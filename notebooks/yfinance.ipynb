```python
import pandas as pd
import yfinance as yf

data = yf.download(["AAPL", "GOOG","IBM"], start = "2024-01-01", end = "2025-12-31")
print(data["Close"].head()) #only show Closing prices

ticker = yf.Ticker("AAPL")
info = ticker.info
print(info["sector"], info["longBusinessSummary"])
print(ticker.earnings)

price = ticker.history(period = "1d")["Close"].iloc[-1]
print(f"Current price: {price}")
```

```python
import pandas as pd
import yfinance as yf
data = {
    "Date: " : ["2020-01-01", "2020-01-02"],
    "Headline: " : [
        "Apple stock surges after tarrifs", "China tarrifs hit tech"
    ],
    "Sentiment: " : [-0.6, 0.1]
}

df = pd.DataFrame(data)

df["Signal"] = df["Sentiment"].apply(lambda x: "Buy" if x > 4 else "Sell" if x < -0.3 else "Neutral")

positive_df = df[df["Signal"] == "Buy"]

df.to_csv("news_signals.csv", index = False)
```
